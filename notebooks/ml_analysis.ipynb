```python
import sys
import os

# Add project root to path
ROOT_DIR = os.path.dirname(os.path.abspath(''))
sys.path.append(ROOT_DIR)

from models.rf_model import train_random_forest
from models.xgboost_model import train_xgboost
from models.svm_model import train_svm
```

```python
rf_acc = train_random_forest()
xgb_acc = train_xgboost()
svm_acc = train_svm()
```

```python
print(f"Random Forest Accuracy: {rf_acc}")
print(f"XGBoost Accuracy: {xgb_acc}")
print(f"SVM Accuracy: {svm_acc}")
```

```python
import matplotlib.pyplot as plt

models = ['Random Forest', 'XGBoost', 'SVM']
accuracies = [rf_acc, xgb_acc, svm_acc]

plt.figure(figsize=(8,5))
plt.bar(models, accuracies)
plt.title("Accuracy Comparison of ML Models")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    "Model": models,
    "Accuracy": accuracies
})

# Ensure the output directory exists (ROOT_DIR and os were defined/imported in an earlier cell)
output_dir = os.path.join(ROOT_DIR, "results")
os.makedirs(output_dir, exist_ok=True)

output_path = os.path.join(output_dir, "Model_Accuracies.csv")
df.to_csv(output_path, index=False)
df
```
